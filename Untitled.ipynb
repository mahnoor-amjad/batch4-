```python
import pandas as pd
```

```python
df = pd.read_excel("student data.csv.xlsx")
```

```python
def assign_grade(score):
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
```

```python
df['Grade'] = df['Percentage'].apply(assign_grade)
```

```python
df.to_csv("student_data_with_grades.csv", index=False)
```

```python
grade_summary = df['Grade'].value_counts().sort_index()
print("Grade Distribution:\n", grade_summary)
```
