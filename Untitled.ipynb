```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv('MhZsQF.csv')
```

```python
df.dtypes
```

```python
df.drop(columns = ['Latitude','Longtitude'] , inplace=True)
```

```python
df.head()
```

```python
df.rename(columns ={'City Code' : 'City','Order Priority Code' : 'Order Priority'} , inplace=True)
df
```

```python
df[df.duplicated()]
```

```python
df=df.drop_duplicates()
df
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df['Total'] = df['quantity'].mul(df['price'])
df
```

```python
df['Sales'] = df['price'] * df['quantity'].fillna((df['price'] * df['quantity']).median())
df
```

```python
df['Sales'] = df['Sales'].fillna(df['Sales'].median())
df
```

```python
df["date"] = pd.to_datetime(df["date"], dayfirst=True)
```

```python
top_products = df.groupby("product")["Sales"].sum().sort_values(ascending=False).head(5)
top_products
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top 5 Best Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_counts = df['payment_method'].value_counts()
payment_counts
```

```python
plt.figure(figsize=(7,7))
plt.pie(payment_counts, labels=payment_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Payment Method Distribution")
plt.show()
```

```python
month_sales = df.groupby("Month")["total"].sum()
month_sales
```

```python
plt.figure(figsize=(10,5))
plt.plot(month_sales.index, month_sales.values, marker="o")
plt.title("Month-wise Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
city_map = {
    1: "Karachi",
    2: "Lahore",
    3: "Islamabad",
    4: "Rawalpindi",
    5: "Multan"
}

df["City"] = df["City"].map(city_map)

city_sales = df.groupby("City")["total"].sum().sort_values()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(y=city_sales.index, x=city_sales.values)
plt.title("City-wise Sales")
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.show()
```

```python
category_month_sales = df.groupby(["Month", "category"])["total"].sum().reset_index()
```

```python
pivot_data = category_month_sales.pivot(index="Month", columns="category", values="total")
```

```python
pivot_data.plot(kind="bar", figsize=(12,6))
plt.title("Monthly Sales by Category")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```
