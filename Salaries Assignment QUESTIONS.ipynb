Drop rows where EmployeeName and JobTitle are 'Not provided'

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

Change BasePay	OvertimePay	OtherPay	Benefits type to float

Display all employees name and basepay columns who have a BasePay greater than 200,000.

Find all employees whose JobTitle contains the word “Manager” (case = False)

Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.

Find count of employees who have zero OvertimePay.

Calculate the TotalPay for each JobTitle where totalpay > 300000

Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order

Replace all missing Benefits values with 0

Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe

Find the number of employees per JobTitle.

```python
import pandas as pd
```

```python
df = pd.read_excel("Salaries.xlsx")
```

```python
df = df[~((df['EmployeeName'] == 'Not provided') & (df['JobTitle'] == 'Not provided'))]
```

```python
cols_to_float = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']
df[cols_to_float] = df[cols_to_float].apply(pd.to_numeric, errors='coerce')
```

```python
high_basepay = df.loc[df['BasePay'] > 200000, ['EmployeeName', 'BasePay']]
print("Employees with BasePay > 200,000:\n", high_basepay.head())
```

```python
managers = df[df['JobTitle'].str.contains('Manager', case=False, na=False)]
print("\nEmployees with 'Manager' in JobTitle:\n", managers.head())
```

```python
selected_columns = df[['EmployeeName', 'JobTitle', 'TotalPay']]
print("\nSelected Columns:\n", selected_columns.head())
```

```python
zero_overtime_count = df[df['OvertimePay'] == 0].shape[0]
print("\nCount of employees with zero OvertimePay:", zero_overtime_count)
```

```python
jobtitle_totalpay = df[df['TotalPay'] > 300000].groupby('JobTitle')['TotalPay'].sum().reset_index()
print("\nTotalPay (JobTitles where TotalPay > 300000):\n", jobtitle_totalpay.head())
```

```python
top5_avg_totalpay = df.groupby('JobTitle')['TotalPayBenefits'].mean().nlargest(5).reset_index()
print("\nTop 5 JobTitles with highest Avg TotalPayBenefits:\n", top5_avg_totalpay)
```

```python
df.fillna({'Benefits': 0}, inplace=True)
```

```python
top_employee_per_year = df.loc[df.groupby('Year')['TotalPay'].idxmax(), ['Year', 'EmployeeName', 'TotalPay']]
print("\nTop Employee by TotalPay in each Year:\n", top_employee_per_year)
```

```python
employee_count_per_job = df['JobTitle'].value_counts().reset_index()
employee_count_per_job.columns = ['JobTitle', 'EmployeeCount']
print("\nNumber of employees per JobTitle:\n", employee_count_per_job.head())
```
